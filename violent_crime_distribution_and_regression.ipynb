```python
from __future__ import division
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
# import mpld3
```

```python
df = pd.DataFrame.from_csv('Crimes_-_2001_to_present.csv', infer_datetime_format=True)
```

```python
df['Date Only'] = pd.to_datetime(df['Date'], format="%m/%d/%Y", exact=False)
```

```python
df['Year'] = df['Date Only'].dt.year
```

```python
df['Primary Type'].value_counts()
```

```python
df.head()
```

```python
battery_cnt = df[(df['Primary Type'] == 'BATTERY')][['Date Only', 'Primary Type', 'Year']]
```

```python
sns.set() #reset parameters so they dont layer on top of each other every time you change
sns.set_palette("pastel")
sns.set_style("darkgrid")
sns.set(font_scale=1.5, palette="Set1")


mpl.rc("figure", figsize=(16, 8))
# font = {'family' : 'normal',
#         'weight' : 'bold',
#         'size'   : 22}
# mpl.rc('font', **font)
```

```python
battery_cnt['Date Only'].value_counts()['2008':'2011'].plot()
plt.title('Batteries Per Day - 2010')
```

```python
battery_cnt['Date Only'].value_counts()['2010'].plot()
plt.title('Batteries Per Day - 2010')
```
